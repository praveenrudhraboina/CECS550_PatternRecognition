## Part 5

1. Use your array of random row indices to create a new matrix `T` from `identity_matrix`. Overall, `T` has the same elements as `identity_matrix`, but the order of the rows of `T` should be according to the variable `rows`
2. Take a moment to think about what we are doing here. Describe in your own words what matrix `T` is.
3. Sanity check, print the shape of `T` and make sure it is the same size as `identity_matrix` and `img_average_channel`. If you need to,  reshape it to have the same size.


```python
import numpy as np

# 'img_average_channel' is your image matrix and 'rows' contains shuffled row indices
num_rows = img_single_channel.shape[0]
identity_matrix = np.eye(num_rows)

# Matrix T by rearranging rows of identity_matrix based on 'rows'
T = identity_matrix[rows]
```

**T** The code rearranges the rows of identity_matrix to create a new matrix T, where the rows are in a different order compared to identity_matrix. This rearrangement is achieved without explicitly using permutation operations, but rather by selecting rows in a specific order.

```python
# check: Print the shape of T
print("Shape of T:", T.shape)

# T has same size as identity_matrix and img_average_channel 
# Reshaping if needed
if T.shape != identity_matrix.shape:
    T = T.reshape(identity_matrix.shape) 

print("Shape of T after reshaping:", T.shape)
```

## Part 7

1. We now want to retrieve the original image. Is matrix `T` invertible? How would you show that it is invertible?
2. Compute the inverse and store the matrix in `Tinv`.
3. Perform the inverse of the scrambling operation and plot the result to show that you retrieved the original image.

```python
import numpy as np
import matplotlib.pyplot as plt

# Checking if matrix 'T' is invertible by computing the matrix determinant
determinant_T = np.linalg.det(T)

if determinant_T != 0:
    # Computing the inverse of 'T'
    Tinv = np.linalg.inv(T)

    # Performing the inverse scrambling operation to retrieve original image
    original_image_retrieved = Tinv @ transformed_image

    # Showing Both original and retrieved images side by side
    plt.figure(figsize=(12, 5)) 
    print("Matrix 'T' is invertible")
    plt.subplot(1, 2, 1)
    plt.imshow(original_image_retrieved)  
    plt.title("Retrieved Original Image")

    plt.subplot(1, 2, 2)
    plt.imshow(img_average_channel) 
    plt.title("Original Image")
    
    plt.show()
    
else:
    print("Matrix 'T' is not invertible, and the original image cannot be retrieved.")
```

## Extra Points

* Now can you perform the scrambling on __both the columns and rows__ of the original image?  Write the code and plot the scrambled image. Then invert the operations and show that you can recover the original image. 

```python
# Create an aray using the image
image = np.array(Image.open('../../Images/lemon2.jpeg'))

# Get height, width and channels of the image
height, width, channels = image.shape

# Create and identity matrice for both rows and columns
identity_rows, identity_cols = np.eye(height), np.eye(width)

# Shuffle both row and column indices
row_indices, col_indices = np.linspace(0, height - 1, height, dtype=int), np.linspace(0, height - 1, width, dtype=int)

np.random.shuffle(row_indices)
np.random.shuffle(col_indices)


# Create a variable with same dimensions as image
scrambled_image = np.zeros_like(image)

# Create matrices T_rows and T_cols for row and column shuffling
T_rows, T_cols = identity_rows[row_indices, :], identity_cols[:, col_indices]

# Scramble the channels 
for channel in range(channels):
    scrambled_image[:, :, channel] = np.dot(T_rows, np.dot(image[:, :, channel], T_cols))

# Plotting scrambled image
plt.imshow(scrambled_image.astype(np.uint8)); plt.axis('off'); plt.title("Scrambled Image"); plt.show()

T_cols_inv, T_rows_inv = np.linalg.inv(T_cols), np.linalg.inv(T_rows)
recovered_image = np.zeros_like(image)

# Retrieving  original image 
for channel in range(channels):
    recovered_image[:, :, channel] = np.dot(T_rows_inv, np.dot(scrambled_image[:, :, channel], T_cols_inv))
# Plot the retrieved original image
plt.imshow(recovered_image); plt.title("Retrieved Original Image"); plt.axis('off'); plt.show()
```
